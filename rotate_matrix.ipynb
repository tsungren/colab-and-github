```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import math as m
from sympy.abc import a, b, c, x, y, z, i, j, k
```

```python
RoA = sp.Matrix([[1,0, 0],
                 [0, sp.cos(a), -sp.sin(a)],
                 [0, sp.sin(a), sp.cos(a)]])

RoB = sp.Matrix([[sp.cos(b),0, sp.sin(b)],
                 [0, 1, 0],
                 [-sp.sin(b), 0, sp.cos(b)]])

RoC = sp.Matrix([[sp.cos(c), -sp.sin(c), 0],
                 [sp.sin(c), sp.cos(c), 0],
                 [0, 0, 1]])
```

```python
roAC = RoC*RoA
roBC = RoC*RoB
```

```python
toolvec = ([2, 5, 9])
uni_i = toolvec[0]/(m.sqrt(toolvec[0]**2+toolvec[1]**2+toolvec[2]**2))
uni_j = toolvec[1]/(m.sqrt(toolvec[0]**2+toolvec[1]**2+toolvec[2]**2))
uni_k = toolvec[2]/(m.sqrt(toolvec[0]**2+toolvec[1]**2+toolvec[2]**2))


radA1 = m.acos(uni_k) #A角第1解的弳度值
radA2 = -m.acos(uni_k) #A角第2解的弳度值

#radC1 = m.atan2(i/m.sin(radA1), -j/m.sin(radA1)) #C角第1解的弳度值
#radC2 = m.atan2(i/m.sin(radA2), -j/m.sin(radA2)) #C角第2解的弳度值
C_ac = sp.atan2((i/sp.sin(a)), -j/sp.sin(a))
```

```python
C_ac
```

```python
radA1
```

```python
deg_C1_ac=sp.N(C_ac.subs({a:radA1, i:uni_i, j:uni_j}))
deg_C2_ac=sp.N(C_ac.subs({a:radA2, i:uni_i, j:uni_j}))
```

```python
m.degrees(deg_C1_ac)
```
